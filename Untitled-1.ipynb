```python
import pandas as pd

# -----------------------------------------
# Sample DataFrames for Concatenation
# -----------------------------------------

df1 = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
})

df2 = pd.DataFrame({
    "A": ["A4", "A5", "A6", "A7"],
    "B": ["B4", "B5", "B6", "B7"],
    "C": ["C4", "C5", "C6", "C7"],
    "D": ["D4", "D5", "D6", "D7"],
})

df3 = pd.DataFrame({
    "A": ["A8", "A9", "A10", "A11"],
    "B": ["B8", "B9", "B10", "B11"],
    "C": ["C8", "C9", "C10", "C11"],
    "D": ["D8", "D9", "D10", "D11"],
})

# -----------------------------------------
# Combine DataFrames into a list
# -----------------------------------------

frames = [df1, df2, df3]

# -----------------------------------------
# 1. Concatenate with ignore_index=True
# Creates a clean 0 → n index
# -----------------------------------------

result_ignore_index = pd.concat(frames, ignore_index=True)
result_ignore_index

# -----------------------------------------
# 2. Concatenate with keys
# Creates a MultiIndex to track original DF
# -----------------------------------------

result_keys = pd.concat(frames, keys=['x', 'y', 'z'])
result_keys

# -----------------------------------------
# 3. Extract one of the original DataFrames
# Using the MultiIndex level
# -----------------------------------------

result_y = result_keys.loc['y']
result_y
```

Key ParametersTo customize your merge in Jupyter, use these parameters:how: Defines the type of join. The default is 'inner'.'inner': Keeps only rows where the key exists in both DataFrames.'left': Keeps all rows from the left DataFrame and matches from the right.'right': Keeps all rows from the right DataFrame and matches from the left.'outer': Keeps all rows from both DataFrames, filling missing values with NaN.on: The common column name(s) to join on. Use a list ['col1', 'col2'] for multiple columns.left_on & right_on: Use these if the joining columns have different names in each DataFrame.suffixes: A tuple to rename overlapping columns that are not keys (e.g., suffixes=('_left', '_right')).indicator: Set to True to add a _merge column showing whether each row came from 'left_only', 'right_only', or 'both'.

import pandas as pd

right = pd.DataFrame({

Notes:

how="inner" keeps ONLY matching rows

Only rows where left.B == right.B appear

Since left has only one row with B=2, it matches all right rows with B=2

Pandas automatically renames duplicate column names:

A_x = from left

A_y = from right

result = pd.merge(left, right, on="B", how="inner")

import pandas as pd

# ---------------------------------------------------------
# SETUP: Example DataFrames
# ---------------------------------------------------------
# df1 and df2 share some matching index values (r1, r3)
# df2 has extra rows (r4) that df1 does not have

df1 = pd.DataFrame({
    "Courses": ["Spark", "Java", "Python"],
    "Fee": [20000, 25000, 22000],
    "Duration": ["30days", "40days", "35days"]
}, index=["r1", "r2", "r3"])

df2 = pd.DataFrame({
    "Courses": ["Spark", "Java", "Python", "Go"],
    "Discount": [2000, 2300, 1200, 2000]
}, index=["r1", "r2", "r3", "r4"])

# ---------------------------------------------------------
# 1. INNER JOIN (only matching index values)
# ---------------------------------------------------------
# Keeps rows where BOTH DataFrames share the same index.
# Adds suffixes to avoid duplicate column names.

df3 = df1.join(df2, lsuffix="_left", rsuffix="_right", how="inner")
print("INNER JOIN RESULT:")
print(df3)

# Expected Output:
#             Courses_left   Fee   Duration   Courses_right   Discount
# r1              Spark     20000   30days        Spark         2000
# r3             Python     22000   35days       Python         1200


# ---------------------------------------------------------
# 2. RIGHT JOIN (keeps all rows from df2)
# ---------------------------------------------------------
# Keeps ALL rows from the right DataFrame (df2).
# Missing values from df1 become NaN.

df3 = df1.join(df2, lsuffix="_left", rsuffix="_right", how="right")
print("\nRIGHT JOIN RESULT:")
print(df3)

# Expected Output:
#             Courses_left     Fee   Duration   Courses_right   Discount
# r1              Spark     20000.0   30days       Spark         2000
# r2               NaN         NaN      NaN        Java          2300
# r3             Python     22000.0   35days      Python         1200
# r4               NaN         NaN      NaN         Go           2000


# ---------------------------------------------------------
# 3. JOIN USING A COLUMN AS THE INDEX
# ---------------------------------------------------------
# .join() normally uses the index.
# To join on a column, set that column as the index first.
# This is similar to SQL join on a key column.

df3 = df1.set_index("Courses").join(
    df2.set_index("Courses"),
    how="inner"
)
print("\nJOIN ON COLUMN RESULT:")
print(df3)

# Expected Output:
#           Fee   Duration   Discount
# Spark    20000   30days     2000
# Python   22000   35days     1200
